```python
data_path = '~/jyh/verl/data/openr1.parquet'
```

输出所有的列名和数据源

```python
df = pd.read_parquet(data_path, engine='pyarrow')
print(df.columns)
data_source_set = set()
for i in df['data_source']:
    data_source_set.add(i)
print(data_source_set)
```

```python
ground_truth_type_set = set()
for i in df['reward_model']:
    ground_truth_type_set.add(type(i['ground_truth']))
print(ground_truth_type_set)
print(len(df['reward_model']))
```
